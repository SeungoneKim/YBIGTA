# Word Cloud

For the corpus to visualize my wordcloud, I chose the Brown corpus from NLTK package.
Specifically, I chose the science_fiction genre within the Brown corpus.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline 
```

```python
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
import nltk
from nltk.corpus import brown

nltk.download('brown')
```

```python
sf_text = brown.words(categories='science_fiction')
sf_text = ' '.join(sf_text)
```

```python
sfwordcloud = WordCloud(background_color='purple',collocations=False,stopwords=stopwords)
```

```python
sfwordcloud.generate(sf_text)
```

```python
sfwordcloud.words_
```

```python
sfwordcloud.to_image()
```

```python
from PIL import Image, ImageFilter
```

```python
mask_arr = np.array(Image.open('data/alien.png'))
mask_arr
```

```python
sfwordcloud = WordCloud(background_color = 'white', collocations=False, stopwords=stopwords,
                       mask = mask_arr, width=400, height=400)
sfwordcloud.generate(sf_text)
sfwordcloud.to_image()
```

```python
darker_sfwordcloud = WordCloud(background_color = 'black', collocations=False, stopwords=stopwords,
                       mask = mask_arr, width=400, height=400, colormap='copper')
```

```python
darker_sfwordcloud.generate(sf_text)
darker_sfwordcloud.to_image()
```
